```python
df['Income_Loan_Ratio'] = df['Total_Income'] / df['LoanAmount']
```

```python
loan_counts= df['Loan_Status'].value_counts()
```

```python
df.columns
```

```python
def risk_category(x):

    if x > 60:
        return "Low Risk"

    elif x > 30:
        return "Medium Risk"

    else:
        return "High Risk"

df['Risk_Category'] = df['Income_Loan_Ratio'].apply(risk_category) 
```

```python
risk_counts = df['Risk_Category'].value_counts()

plt.bar(risk_counts.index, risk_counts.values)

plt.title("Loan Risk Distribution")

plt.show()
```

```python
risk_status = df.groupby(['Risk_Category','Loan_Status']).size().unstack()

risk_status.plot(kind='bar')

plt.title("Risk Category vs Loan Approval")

plt.show()
```

```python
# Medium Risk applicants have higher approval probability.
```

```python
df['Risk_Score'] = 0

df.loc[df['Credit_History']==1, 'Risk_Score'] += 50

df.loc[df['Income_Loan_Ratio']>60, 'Risk_Score'] += 20

df.loc[df['Education']=='Graduate', 'Risk_Score'] += 10

df.loc[df['Property_Area']=='Urban', 'Risk_Score'] += 10
```

```python
df['Predicted_Loan_Status'] = df['Risk_Score'].apply(lambda x: "Y" if x>=60 else "N")
```

```python
pred_counts = df['Predicted_Loan_Status'].value_counts()

plt.bar(pred_counts.index, pred_counts.values)

plt.title("Predicted Loan Approval")

plt.show()
```

```python
accuracy = (df['Loan_Status'] == df['Predicted_Loan_Status']).mean()

print("Prediction Accuracy:", accuracy)
```

```python
df['Estimated_Interest'] = df['LoanAmount'] * 0.12

df['Expected_Loss'] = np.where(df['Risk_Category']=="High Risk",
                               df['LoanAmount']*0.5,
                               0)

df['Estimated_Profit'] = df['Estimated_Interest'] - df['Expected_Loss']
```

```python
plt.hist(df['Estimated_Profit'], bins=30)

plt.title("Loan Profitability Distribution")

plt.xlabel("Profit")

plt.show()
```
